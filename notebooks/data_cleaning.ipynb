```python
import pandas as pd
```

```python
ruta = "../../data/tviaje.csv"  # Ruta externa por el peso del archivo

data = pd.read_csv(ruta)
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
important_vars = [
    "p5_9_1", # hora comienzo
    "p5_9_2", # minuto comienzo
    "p5_10_1", # hora fin
    "p5_10_2", # minuto fin
    "p5_14_01",
    "p5_14_02",
    "p5_14_03",
    "p5_14_04",
    "p5_14_05",
    "p5_14_07",
    "p5_14_09",
    "p5_14_14",
    "p5_7_7",
    "p5_12_7",
    "p5_6",
    "p5_11a",
    "estrato",
]

data_clean = data[important_vars].copy()
data_clean.shape
```

```python
data_clean["tiempo"] = (
    (data_clean["p5_10_1"] * 60 + data_clean["p5_10_2"]) -
    (data_clean["p5_9_1"] * 60 + data_clean["p5_9_2"])
)
# Para los que quedan en negativos (llegan el dia siguiente)
data_clean["tiempo"] = data_clean["tiempo"].where(data_clean["tiempo"] >= 0, data_clean["tiempo"] + 24*60)
data_clean = data_clean[data_clean["tiempo"] != 0]
```

```python
data_clean.describe()
```

```python
data_clean.drop(columns=["p5_9_1", "p5_9_2", "p5_10_1", "p5_10_2"], inplace=True)
```

```python
mapa_columnas = {
    "p5_14_01": "auto",
    "p5_14_02": "colectivo",
    "p5_14_03": "uber_etc",
    "p5_14_04": "taxi",
    "p5_14_05": "metro",
    "p5_14_07": "bici",
    "p5_14_09": "moto",
    "p5_14_14": "caminar",
    "p5_7_7": "entidad_origen",
    "p5_12_7": "entidad_destino",
    "p5_6": "tipo_origen",
    "p5_11a": "tipo_destino"
}

data_clean = data_clean.rename(columns=mapa_columnas)
```

```python
data_clean.head()
```

## Discretizar tiempo

```python
data_clean.describe()
```

Discretizacion por cuantiles con pd.qcut:

```python
data_clean['tiempo_bin'] = pd.qcut(data_clean['tiempo'], q=10, labels=None, duplicates='drop')
print(data_clean['tiempo_bin'].cat.categories)
```

```python
data_clean['tiempo_bin'] = data_clean['tiempo_bin'].astype(str)
data_clean = data_clean.drop(columns=['tiempo'])
print(data_clean['tiempo_bin'].unique())
```
